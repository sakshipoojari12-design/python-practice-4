```python
#accept project marks,internal marks,external marks from student
#passmarks is 50 in each subject
#calculate total score if student got more than or equal to 50 in each subject
#total score=70% on project+20% on external+ 10% on internal
#print A grade if score is 90+
#print B grade if score IS B/W 70-90
#print C grade if score is B/W 50-70
project=int(input("Enter project marks:"))     #50
internal=int(input("Enter internal marks:"))   # 50
external=int(input("Enter external marks:"))   # 50
if project>=50 and internal>=50 and external>=50:
    total=(project*70)/100+ (external*20)/100+(internal*10)/100
    print("Total score is: ",total)
    if total>=90:
        print("A grade")
    elif total>=70:
        print("B grade")
    else:
        print("C grade")

else:
    print("fail")
```

```python
# Write a program to check given number is prime or not
# 2 3 5 7 11 13 17 19 ....
count=0
num=int(input("Enter your number to check it is prime or not :"))
for i in range(1, (num//2)+1):
    if num%i==0:
        count=count+1

if count==1:
    print(num," is a prime number")
else:
    print(num," is not a prime number")
```

```python
# Write a program to check given number is perfect number or not
# 6      :  1+2+3  ==6
# 28     : 1+2+4+7+14===28
# 10     : 1+2+5  ===> 8

sum=0
num=int(input("Enter your number to check it is perfect number or not :"))
for i in range(1, (num//2)+1):
    if num%i==0:
        sum=sum+i

if sum==num:
    print(num," is perfect number")
else:
    print(num," is not a perfect number")
```

```python
# Write a program to find number of digits in a given number
 #input: 5235
 #op: 4

num=6574
count=0
while num!=0:           # 0 !=0
    num=num//10         # 6//10  --> 0
    count=count+1        # 4
print("Number of digits : ",count)
```

```python
# Write a program to display the number in reverse order
# input: 6324  ---
# output: 4236

num=6324
rev=0
while num!=0:            # 0!=0
    rem=num%10               # 6%10   =6
    rev=rev*10+rem       #4236
    num=num//10              # 6//10  0
print("reverse number: ",rev)
```
